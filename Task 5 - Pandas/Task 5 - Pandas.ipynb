**Data Selection â€“ Based on Conditional filtering**

```python
print("Movies with 8.5+ Rating")
moviesWith8point5PlusRating = df.loc[(df["Rating"] > 8.5), ["Title", "Director", "Rating", "Year"]]
moviesWith8point5PlusRating
```

```python
print("Movies with 8.5+ Rating directed by Christopher Nolan")
moviesWith8point5PlusRating.loc[(moviesWith8point5PlusRating["Director"] == "Christopher Nolan"), ["Title", "Rating"]]
```

```python
print("Movies released in 2011 with a rating of 8+")
df[(df["Year"] == 2011) & (df["Rating"] > 8)]
```

___

**Groupby operations**

```python
df.groupby("Director")["Rating"].mean()
```

```python
print("Number of movies released in each year")
df.groupby("Year")["Year"].count()
```

___

**Sorting operation**

```python
print("Number of movies in each year without Sorting")
df.groupby("Year", sort = False)["Year"].count()
```

```python
print("Movies with 8.5+ Rating sorted by Year of release")
moviesWith8point5PlusRating.sort_values("Year")
```
